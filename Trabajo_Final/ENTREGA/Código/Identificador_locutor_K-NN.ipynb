```python
from sklearn.model_selection import train_test_split

# Creamos la matriz de entrenamiento xtrain con los valores del pitch
xtrain = [[x11, x12],
          [x21, x22],
          [x31, x32],
          [x41, x42]
         ]

# Creamos la matriz de la clase asociada a cada conjunto de datos, ytrain
ytrain = [['loc1'],
          ['loc2'],
          ['loc3'],
          ['loc4']
         ]

# Visualizamos los datos de entrenamiento
print('------TRAIN--------')
print(xtrain)
print(ytrain)
print('-------------------')

# Creamos la matriz de test a partir de los datos de las matrices de entrenamiento.
# Se utilizan el 30% de los datos de las matrices de entrenamiento para crear la
# matriz de test.
x, xtest, y, ytest = train_test_split(xtrain, ytrain, test_size=0.3)

# Visualizamos los datos de test
print('------TEST-------')
print(xtest)
print(ytest)
print('-----------------')


```

Cargar datos para la aplicacion del algoritmo K-NN

```python
from sklearn.neighbors import KNeighborsClassifier
import sys

# Creamos varios clasificadores para ver con cuantos n vecinos mas proximos es mas preciso

# Creamos el clasificador para n=1 
KNN_classifier = KNeighborsClassifier(n_neighbors=1)

# Entrenamiento
KNN_classifier.fit(xtrain, ytrain)

# Prediccion
y_hat_KNN = KNN_classifier.predict(xtest)
print(y_hat_KNN)

# Precision
acc = np.sum (ytest == y_hat_KNN)/len(ytest)
print('El algoritmo con n=1 tiene una precision del: ' + str(acc))


# Creamos el clasificador para n=2
KNN_classifier = KNeighborsClassifier(n_neighbors=2)

# Entrenamiento
KNN_classifier.fit(xtrain, ytrain)

# Prediccion
y_hat_KNN = KNN_classifier.predict(xtest)
print(y_hat_KNN)

# Precision
acc = np.sum (ytest == y_hat_KNN)/len(ytest)
print('El algoritmo con n=2 tiene una precision del: ' + str(acc))


# Creamos el clasificador para n=3
KNN_classifier = KNeighborsClassifier(n_neighbors=3)

# Entrenamiento
KNN_classifier.fit(xtrain, ytrain)

# Prediccion
y_hat_KNN = KNN_classifier.predict(xtest)
print(y_hat_KNN)

# Precision
acc = np.sum (ytest == y_hat_KNN)/len(ytest)
print('El algoritmo con n=3 tiene una precision del: ' + str(acc))


# Creamos el clasificador para n=4
KNN_classifier = KNeighborsClassifier(n_neighbors=4)

# Entrenamiento
KNN_classifier.fit(xtrain, ytrain)

# Prediccion
y_hat_KNN = KNN_classifier.predict(xtest)
print(y_hat_KNN)

# Precision
acc = np.sum (ytest == y_hat_KNN)/len(ytest)
print('El algoritmo con n=4 tiene una precision del: ' + str(acc))
```
